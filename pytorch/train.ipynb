```python
def transform_string(batch):
    for line in batch:
        line = re.sub(' +', ' ', line.strip())  # 띄어쓰기 2개있는 문장 제거        
        inputs = [vocab(list(char)) for char in line]
        padded_inputs = torch.nn.utils.rnn.pad_sequence(inputs, dtype=torch.int32)
```

```python
with open('./namuwikitext_20200302.dev', 'r', encoding='utf-8') as f:
    train_ds = f.readlines()
    dataloader = DataLoader(train_ds, batch_size=8, shuffle=False, collate_fn=transform_string)
```
